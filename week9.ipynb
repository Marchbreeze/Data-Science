## 1. k-NN 기반 분류

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
# 지뢰와 보물 데이터 불러오기
df = pd.read_csv('treasure.csv')
df
```

```python
# 주어진 데이터에서 입력 변수와 예측 변수를 분리
df_x = df[['horizontal', 'vertical']]
df_y = df['MineOrTreasure']
```

```python
# 데이터를 학습용과 검증용으로 분리
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.3, random_state=0)
```

```python
# 학습용 데이터를 바탕으로 k-NN 분류기 생성
kNN = KNeighborsClassifier(n_neighbors = 3, metric = 'euclidean')
kNN.fit(x_train, y_train)
```

```python
# 만들어진 k-NN 분류기를 검증용 데이터에 적용
y_pred = kNN.predict(x_test)
y_pred
```

```python
# Confusion Matrix 출력
confusion_matrix(y_test, y_pred, labels = ['MINE', 'TREASURE'])
```

## 2. NB

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
df = pd.read_csv('baseball.csv')
df
```

```python
# 주어진 데이터에서 입력 변수와 예측 변수를 분리
df_x = df[['error', 'pitcher', 'score', 'homerun']]
df_y = df['GameResult']
```

```python
# NB 분류기 생성
NB = BernoulliNB(alpha = 0)
NB.fit(df_x, df_y)
```

```python
# 새로운 경기 데이터 제공
x_test = pd.read_csv('ku_vs_monsters.csv')
x_test
```

```python
# NB 분류기를 새로운 경기 데이터에 적용
y_pred = NB.predict(x_test)
y_pred
```

## 3. 로지스틱 회귀

```python
# 데이터 불러오고 독립, 종속 변수 구분
df = pd.read_csv('telecom_logistic.csv')
df_x = df.iloc[:, 0:7]
df_y = df.iloc[:, 7]
```

```python
# 로지스틱 회귀 모형 생성
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(df_x, df_y)

print(LR.coef_)
print(LR.intercept_)
```
